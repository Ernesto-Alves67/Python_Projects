# Quais paises tem mais veiculos de guerra(terrestres)?

```python
df.loc[df['Tanks'] < 1000]
anw = df[['Tanks', 'Tanker Fleet', 'country', 'Armored Vehicles']].sort_values('Tanks', ascending=False)
anw.head(20)
```

# Quais paises tem mais veiculos aéreos?

```python
anw = df[['Attack Helicopters', 'Helicopter Carriers', 'Helicopters', 'country']].sort_values('Attack Helicopters', ascending=False)
anw.head(10)
```

# Quais paises tem mais veiculos acquaticos?

```python
df['Total Acgua'] = df['Frigates'] + df['Destroyers']
anw = df[['country','Frigates', 'Destroyers','Total Acgua']]
anw.sort_values('Total Acgua', ascending=False).head(10)
```

# Quais paises mais investem em 'Defense Budget'?

```python
anw=df[['Defense Budget', 'country']].sort_values('Defense Budget', ascending=False)
anw.head(10)
```

# Quais paises tem a menor e a maior divida externa?

```python
anw=df[['Extrnal Debt', 'country']].sort_values('External Debt', ascending=False)
anw.head(20)
```

# Quais paises consomem mais petróleo? E quais produzem menos?

```python
anw = df[['Oil Production', 'Oil Consumption', 'country']]
rC = anw.sort_values('Oil Consumption', ascending = False).head(10)
rP = anw.sort_values('Oil Production', ascending=False).head(10)
rC, rP
```

# Quais paises são mais povoados?

```python
anw = df[['Total Population', 'country']]
anw.sort_values('Total Population', ascending=False).head(10)
```

# O povoamento de um pais influencia em seu poder belico?

```python
anw = df[['Total Population', 'country', 'Tanks','Armored Vehicles', 'Available Manpower']]
anw.sort_values('Armored Vehicles', ascending=False).head(10)
```

# Quais paises mais fracos em poder militar?

```python
lista = ['country','Frigates','Tanks', 'Armored Vehicles', 'Destroyers','Corvettes','Aircraft Carriers', 'Navy Ships', 'Fighters/Interceptors']
anw=df[lista]
anw['Total Army'] = anw.loc[anw['Tanks'] > 100, lista[1:]].sum(axis=1)
anw.sort_values('Total Army', ascending=True).head(10)
```

# Qual a situação do poder belico do Brasil?

```python
r1 = df[lista]
r1['Total Army'] = r1.loc[r1['Tanks'] > 0 , lista[1:]].sum(axis=1)
r1.loc[r1['country'] == "Brazil"]
```
